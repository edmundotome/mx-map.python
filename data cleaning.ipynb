```python
import plotly.express as px
import plotly.io as pio
pio.renderers.default = 'browser'
```

```python
mexican_states = json.load(open('mexican_states.geojson','r'))
```

```python
mexican_states['features'][2]['properties']
```

```python
mexican_states['features'][0]['properties']['fips_admin']
```

```python
state_id_map = {}
for feature in mexican_states['features']:
    feature['id'] = feature['properties']['fips_admin']
    state_id_map[feature['properties']['admin_name']] = feature['id']
```

```python
state_id_map
```

```python
df = pd.read_csv('Mexico-crimen-by-state.csv')
```

```python
df
```

```python
df.replace(to_replace='Ciudad de Mexico',value='Distrito Federal',inplace=True)
df.replace(to_replace='MichoacÃ¡n de Ocampo',value='Michoacan',inplace=True)
df.replace(to_replace='Veracruz de Ignacio de la Llave',value='Veracruz',inplace=True)
df.replace(to_replace='Coahuila de Zaragoza',value='Coahuila',inplace=True)
df
```

```python
df['Entidad'] = df['Entidad'].str.normalize('NFKD').str.encode('ascii', errors = 'ignore').str.decode('utf-8')
```

```python
df = df.drop([0])
```

```python
df = df.sort_values('Entidad')
```

```python
df.replace(to_replace='Ciudad de Mexico',value='Distrito Federal',inplace=True)

df
```

```python
df["id"] = df['Entidad'].apply(lambda x: state_id_map[x])
```

```python
fig = px.choropleth(
    df,
    locations="id",
    geojson=state_id_map,
    color='2010',
    center={"lat": 19, "lon": -99}
)
fig.update_geos(fitbounds="locations", visible=False)
fig.show()
```
